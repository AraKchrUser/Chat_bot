```python

html = urlopen('https://www.deviantart.com/djamilaknopf/art/Light-in-the-Dark-689387108?comment=1%3A689387108%3A4878510957')
bs = BeautifulSoup(html, 'html.parser')
```

```python
len(bs.find_all('div', {'data-hook':'comments_thread_item'}))
```

```python
for comm in bs.find_all('div', {'data-hook':'comments_thread_item'}):
    print(comm.find('div', {'data-hook':'comment_body'}).get_text())
```

```python
list(bs.find_all('div', {'data-hook':'comments_thread_item'})\
     [3].find('div', {'data-hook':'comment_body'}))[0].get_text()
```

```python
for comm in bs.find_all('p', {'id':'viewer-foo'}):
    print(comm.span.get_text())
```
